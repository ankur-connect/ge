```python
%matplotlib inline 
import splitter
from tqdm import tqdm

```

```python
splitter.split('mit')
```

```python
with open("/Users/bilals01/Documents/NLP/project/quasi-rnn-run-12-09/dev_source_seqs.txt", "r", encoding="utf8") as f:
    
    allwords = []
    
    for i, line in enumerate(f):
        
        words = line.split()
        for word in words:
            allwords.append(word)
            
allwords_u=[]

for word in allwords: 
    allwords_u.append(word.replace(".",""))
 
```

```python
compounds=[]

for word in tqdm(allwords_u): 
    if len(splitter.split(word,'de_de'))>1 :
        compounds.append(word)
    
```

```python
#len(compounds)
compounds
```

```python
#write the list of compounds in a file
thefile = open('compounds.txt', 'w')

for item in compounds:
  thefile.write("%s\n" % item) 
    
    
```
